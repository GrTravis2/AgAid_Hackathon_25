```python
import matplotlib.pyplot as plt

from sklearn.datasets import make_classification

from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix

from sklearn.model_selection import train_test_split

from sklearn.metrics import jaccard_score

import pandas as pd

import numpy as np

from sklearn.ensemble import RandomForestClassifier
```

```python
image=pd.read_csv('data.csv',header=0, usecols=["y","x","red","green","blue","residue"])

for col in ["y","x","red", "green","blue"]:
  image[col]=pd. to_numeric(image[col], errors='coerce')
image=image.astype({"residue":bool})
image

```

```python
training_data, test_data = train_test_split(image, test_size = 0.25)

training_samples = training_data.drop(columns='residue')
training_labels = training_data['residue']

test_samples = test_data.drop(columns='residue')
test_labels = test_data['residue']
```

```python
rf = RandomForestClassifier(n_estimators=10)
rf.fit(training_samples, training_labels)
```

```python
predict_test=rf.predict(test_samples)
predict_train=rf.predict(training_samples)
```

```python

# For test data
cm_test = confusion_matrix(test_labels, predict_test)
disp_test = ConfusionMatrixDisplay(confusion_matrix=cm_test, display_labels=rf.classes_)
disp_test.plot()
plt.title('Confusion Matrix - Test Data')
plt.show()

# For training data
cm_train = confusion_matrix(training_labels, predict_train)
disp_train = ConfusionMatrixDisplay(confusion_matrix=cm_train, display_labels=rf.classes_)
disp_train.plot()
plt.title('Confusion Matrix - Training Data')
plt.show()


#Calculate and print additional metrics (optional)
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print("Test Data Metrics:")
print(f"Accuracy: {accuracy_score(test_labels, predict_test)}")
print(f"Precision: {precision_score(test_labels, predict_test)}")
print(f"Recall: {recall_score(test_labels, predict_test)}")
print(f"F1-score: {f1_score(test_labels, predict_test)}")

print("\nTraining Data Metrics:")
print(f"Accuracy: {accuracy_score(training_labels, predict_train)}")
print(f"Precision: {precision_score(training_labels, predict_train)}")
print(f"Recall: {recall_score(training_labels, predict_train)}")
print(f"F1-score: {f1_score(training_labels, predict_train)}")
```

```python
print(f"I o U: {jaccard_score(test_labels, predict_test)}")
print(f"I o U: {jaccard_score(training_labels, predict_train)}")
```

```python
print("Test Accuracy:",rf.score(test_samples, test_labels))
print("Train Accuracy:",rf.score(training_samples, training_labels))
```
