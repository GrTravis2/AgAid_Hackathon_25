```python
import matplotlib.pyplot as plt

from sklearn.datasets import make_classification

from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix

from sklearn.model_selection import train_test_split

from sklearn.metrics import jaccard_score

import pandas as pd

import numpy as np

from sklearn.ensemble import RandomForestClassifier
```

```python
test=pd.read_csv('data.csv',header=0,usecols=["y","x","red","green","blue","location","image_num","part_num","residue",])
test
```

```python
test.drop_duplicates(inplace=True)
test.to_csv('data.csv', index=False)
test
```

```python
# prompt: a data frame of only the rows where image_num=939 from test

# Assuming 'test' DataFrame is already loaded as in the provided code.
image_939_df = test[test['image_num'] == 939]
image_939_df.drop(columns=['location', 'image_num', 'part_num'], inplace=True)

for col in ["y","x","red", "green","blue"]:
  image_939_df[col]=pd. to_numeric(image_939_df[col], errors='coerce')
image_939_df=image_939_df.astype({"residue":bool})
image_939_df

```

```python
image=pd.read_csv('data.csv',header=0, usecols=["y","x","red","green","blue","residue"])

for col in ["y","x","red", "green","blue"]:
  image[col]=pd. to_numeric(image[col], errors='coerce')
image=image.astype({"residue":bool})
image

```

```python
training_data, test_data = train_test_split(image, test_size = 0.25)

training_samples = training_data.drop(columns='residue')
training_labels = training_data['residue']

#test_samples = test_data.drop(columns='residue')
#est_labels = test_data['residue']
```

```python
image_939_samples = image_939_df.drop(columns='residue')
image_939_labels = image_939_df['residue']
```

```python
rf = RandomForestClassifier(n_estimators=20)
rf.fit(training_samples, training_labels)
```

```python
#predict_test=rf.predict(test_samples)
predict_train=rf.predict(training_samples)
predict_939=rf.predict(image_939_samples)
```

```python


cm_test = confusion_matrix(image_939_labels, predict_939)
disp_test = ConfusionMatrixDisplay(confusion_matrix=cm_test, display_labels=rf.classes_)
disp_test.plot()
plt.title('Confusion Matrix - Test Data')
plt.show()

# For training data
cm_train = confusion_matrix(training_labels, predict_train)
disp_train = ConfusionMatrixDisplay(confusion_matrix=cm_train, display_labels=rf.classes_)
disp_train.plot()
plt.title('Confusion Matrix - Training Data')
plt.show()


#Calculate and print additional metrics (optional)
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print("Test Data Metrics:")
print(f"Accuracy: {accuracy_score(image_939_labels, predict_939)}")
print(f"Precision: {precision_score(image_939_labels, predict_939)}")
print(f"Recall: {recall_score(image_939_labels, predict_939)}")
print(f"F1-score: {f1_score(image_939_labels, predict_939)}")

print("\nTraining Data Metrics:")
print(f"Accuracy: {accuracy_score(training_labels, predict_train)}")
print(f"Precision: {precision_score(training_labels, predict_train)}")
print(f"Recall: {recall_score(training_labels, predict_train)}")
print(f"F1-score: {f1_score(training_labels, predict_train)}")
```

```python
print(f"I o U: {jaccard_score(image_939_labels, predict_939)}")
print(f"I o U: {jaccard_score(training_labels, predict_train)}")
```

```python
print("Test Accuracy:",rf.score(image_939_samples, image_939_labels))
print("Train Accuracy:",rf.score(training_samples, training_labels))
```
